```python
#1.Create Rectangle  class with attributes length and breadth and methods to find area and perimeter.compare two Rectangle objects by their area


class Rectangle(): #create class rectangle
    def getval(self,length,breadth):
        self.length=length
        self.breadth=breadth
        
       
    def  area(self): #function to calculate area 
        return self.length*self.breadth
 
    def perimeter(self):#function to calculate perimeter  
        return 2*(self.length+self.breadth)
    
    
Rectangle1=Rectangle()#rectangle1 obejct created  

length=(int(input("Enter the Length of Rectangle 1   :")))  #Reading length and breadth
breadth=(int(input("Enter the Breadth of Rectangle 1 :")))

Rectangle1.getval(length,breadth)

print("AREA OF RECTANGLE 1       : ",Rectangle1.area())
print("PERIMETER OF RECTANGLE 1  : ",Rectangle1.perimeter())


Rectangle2=Rectangle()#rectangle1 obejct created  

length=(int(input("\nEnter the Length of Rectangle 2 :")))
breadth=(int(input("Enter the Breadth of Rectangle 2 :")))

Rectangle2.getval(length,breadth)

print("AREA OF RECTANGLE 2       : ",Rectangle2.area())
print("PERIMETER OF RECTANGLE 2  : ",Rectangle2.perimeter())


if(Rectangle2.area()==Rectangle1.area()): #Comparing areas
    print("BOTH HAVE SAME AREA")
else:#
    print("BOTH DONT HAVE SAME AREA")

```

```python
# 4.Create a Bank account with members account number, name, type of account and balance. Write constructor and methods to deposit at
#the bank and withdraw an amount from the bank.


class BankAccount: # class definition

    ' BANK ACCOUNT CLASS DEFINITION '

    def __init__(self,Acc_name,Acc_number,Acc_type):# define constructor
        self.OWNER_NAME = Acc_name
        self.ACCOUNT_NUMBER=Acc_number
        self.ACCOUNT_TYPE = Acc_type         
        self.BALANCE= 0
        print("ACCOUNT CREATED FOR %s WITH ACCOUNT NUMBER %s"%(self.OWNER_NAME,self.ACCOUNT_NUMBER))
        

    def showAccount(self):          # define method
        print("OWNER NAME         : ",self.OWNER_NAME)
        print("ACCOUNT NUMBER     : ",self.ACCOUNT_NUMBER)
        print("ACCOUNT TYPE       : ",self.ACCOUNT_TYPE )
        print("BALANCE            :  %d RS"%self.BALANCE)
    
    def Deposit(self,amount):
        self.BALANCE+=amount
        print("DEPOSITED AMOUNT  : %d RS"%amount)
        print("AVAILABLE BALANCE : %d RS"%self.BALANCE)

    def WithDraw(self,withdraw_amount):
        self.BALANCE-=withdraw_amount
        print("WITHDRAW AMOUNT   : %d RS"%amount)
        print("AVAILABLE BALANCE : %d RS"%self.BALANCE)
        
        
# MAIN PROGRAM

Acc_name   = input("ENTER NAME           : ")
Acc_number = input("ENTER ACCOUNT NUMBER : ")
Acc_type   = input("ENTER ACCOUNT TYPE   : ")

ACCOUNT = BankAccount(Acc_name,Acc_number,Acc_type)#object created

ACCOUNT.showAccount()#showing account details

amount=int(input("ENTER AMOUNT TO DEPOSIT"))
ACCOUNT.Deposit(amount)#amount depostited

amount=int(input("ENTER AMOUNT TO WITHDRAW"))
ACCOUNT.WithDraw(amount)#amount withdrawed



```

```python
# 5. Create a class Publisher (name). Derive class Book from Publisher  with attributes title and author. Derive class Python from Book with
# attributes price and no_of_pages. Write a program that displays  information about a Python book. Use base class constructor
# invocation.

class Publisher: #defining base class 
    def __init__(self,name):
        self.name=name
        
class Book(Publisher):  #defining subclass book
    def __init__(self,name,title,author):
        Publisher.__init__(self,name)
        self.title=title
        self.author=author
        
class Python(Book):#defining subclass python
    def __init__(self,name,title,author,price,no_page):
        Book.__init__(self,name,title,author)
        self.price=price
        self.no_page=no_page
    def display(self):
        print("Name :",self.name)
        print("Title :",self.title)
        print("Author :",self.author)
        print("Price :",self.price)
        print("No.of pages :",self.no_page)
        
book1=Python("Alchemist","The Alchemist","Paulo Coelho","100","163")#created object of subclass python
book1.display() #displaying book details      
    


```

```python
#8 create a class time with private attribute hour,minute and second.overload '+' operator to find sum of 2 time
class time:
    def __init__(self,h,m,s):
        self.hour=h
        self.minute=m
        self.second=s
    def __add__(self,other):
        h=self.hour + other.hour
        m=self.minute + other.minute
        s=self.second + other.second
        if(s>=60):
            m=m+1
            s=s-60
            
        if(m>=60):
            h=h+1
            m=m-60
        return h,m,s
    
print("\t\t\thour,minute,second")
tym1=time(2,40,50)
tym2=time(5,30,30)
add = tym1 + tym2

print("adding time:",add)
```

```python
#Write a Program to implement the following. 
#Create a base class called Person consisting of name and code. Create 2  child classes 
#A) Account with member pay 
#B)Admin with experience 
#Create a class Employee with name, code, experience and pay by  inheriting the above classes.


class Person():
    def __init__(self,name,code):
        self.name=name
        self.code=code
      
    def DisplayPerson(self):
        print("Name : ",self.name)
        print("Code : ",self.code)  
        
class Account(Person):
    def __init__(self,pay):
        self.pay=pay
        
    def DisplayAccount(self):
        print("Payment : ",self.pay)
        
class Admin(Person):
    def __init__(self,experience):
        self.experience=experience
        
    def DisplayAdmin(self):
        print("Experience : ",self.experience)  
        
        
class Employee(Account,Admin):
    def __init__(self,name,code,pay,experience):
        Person.__init__(self,name,code)
        Account.__init__(self,pay)
        Admin.__init__(self,experience)
        
obj=Employee("Ayisha","111","5000","3")

obj.DisplayPerson()
obj.DisplayAccount()
obj.DisplayAdmin()
        
        
```
